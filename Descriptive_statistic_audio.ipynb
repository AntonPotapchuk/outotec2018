```python
import time
from datetime import datetime
import struct
import time
import pandas as pd
import numpy as np

def get_name_from_datetime(datetime1, prefix='device1_channel1_'):
    return audio_dump + '/' + prefix + datetime.strftime(datetime1, '%Y%m%d%H%M%S') + '.wav'

def read_file(path):
    return wave.open(path)

def get_datetime_from_filename(path):
    path = path.split('.')
    sample = '20181013110342'
    return datetime.strptime(path[0][-len(sample) : ], '%Y%m%d%H%M%S')


def read_data(path):
    data = read_file(path)
    start_time = time.time()
    sampwidth = data.getsampwidth()
    fmts = (None, "=B", "=h", None, "=l")
    fmt = fmts[sampwidth]
    dcs  = (None, 128, 0, None, 0)
    dc = dcs[sampwidth]

    size = 1000000
    array = np.zeros(data.getnframes())

    for i in range(data.getnframes()):
        iframe = data.readframes(1)
        iframe = struct.unpack(fmt, iframe)[0]
        iframe -= dc
        array[i] = iframe

    return array

class Status:
    current_time = 0
    
    def __init__(self, quat, acc, audio, audio_start, audio_end):
        self.current_time = max([quat['timestamp'].min(), acc['timestamp'].min(), audio_start])
        self.quat = quat.copy()
        self.quat['timestamp'] = pd.to_datetime(self.quat['timestamp'])
        self.acc = acc
        self.audio = audio
        self.delta_seconds = 600
        self.wait_seconds = 2
        
    def start(self):
        idx = 0
        print(self.current_time)
        self.current_time = datetime.strptime(self.current_time[:len('2018-10-13 15:00:00')], '%Y-%m-%d %H:%M:%S')
        self.current_time_end = self.current_time + pd.Timedelta(seconds=self.delta_seconds)
        while idx < 5:
            time.sleep(self.wait_seconds)
            
            self.plot_quat(self.quat.loc[(self.quat['timestamp'] > self.current_time) & (self.quat['timestamp'] < self.current_time_end)])
            
            
            self.current_time += pd.Timedelta(seconds=self.delta_seconds)
            self.current_time_end += pd.Timedelta(seconds=self.delta_seconds)
            print(self.current_time, self.current_time_end)
            idx += 1
            
    def plot_quat(self, df):
        fig = plt.figure()
        df_pepa = df[df['variable'] == 'NGIMU-PEPA']
        df_pepa.index = pd.to_datetime(df_pepa['timestamp'])
        plt.title('Pepa motion')
        df_pepa['q1'].plot(color='blue', label='pepa_x')
        df_pepa['q2'].plot(color='green', label='pepa_y')
        df_pepa['q3'].plot(color='red', label='pepa_z')
        plt.xlabel('time')
        #plt.show()
        mpld3.save_html(fig, 'temp.html')
        
    def plot_acc(self, df):
        pass
    
motion_dump = "Motion"
audio_dump = "Audio"

data = None
all_files = os.listdir(audio_dump)
for file in os.listdir(audio_dump)[:1]:
    data = wave.open(audio_dump + "/" + file)
array = read_data(audio_dump + "/" + all_files[1])
        
        
quat = pd.read_csv('quat.csv')
acc = pd.read_csv('acc.csv')
status = Status(quat, acc, array, acc['timestamp'].min(), acc['timestamp'].min())
status.start()
```

```python
def get_label(x, df, margin):
    for i in range(len(df.index)):
        if x['time'] > df.loc[df.index[i], 'Time Start'] - margin and x['time'] < df.loc[df.index[i], 'Time Finish'] + margin:
            if df.loc[df.index[i], 'Process status (data label)'] == 'Failure':
                return 1
    return x['label']
            

def get_labels(margin=30, freq='60S'):
    margin = pd.Timedelta(minutes=margin)
    df = pd.read_excel('IHearVoicesData.xlsx')
    df.columns = df.iloc[0]
    df = df.iloc[1:]
    df = df[~df['Time Start'].isin(['SAME STATUS', 'NO DATA'])]
    df['Time Start'] = pd.to_datetime(df['Time Start'])
    df['Time Finish'] = pd.to_datetime(df['Time Finish'])
    labels = pd.DataFrame(index=pd.date_range(start='10/12/2018 13:00', end='10/26/2018 12:00', freq=freq))
    labels['time'] = labels.index
    labels['label'] = labels.apply(lambda x: get_label(x, df, margin), axis=1)
    print(labels.head())
    
get_labels()
    
```

```python
def get_name_from_datetime(datetime1, prefix='device1_channel1_'):
    return audio_dump + '/' + prefix + datetime.strftime(datetime1, '%Y%m%d%H%M%S') + '.wav'

def read_file(path):
    return wave.open(path)

def get_datetime_from_filename(path):
    path = path.split('.')
    sample = '20181013110342'
    return datetime.strptime(path[0][-len(sample) : ], '%Y%m%d%H%M%S')

def get_audio_features_by_datetime(dt, length, all_files):
    start = max([get_datetime_from_filename(file) for file in all_files if get_datetime_from_filename(file) < dt])
    end = min([get_datetime_from_filename(file) for file in all_files if get_datetime_from_filename(file) > dt])
    duration = end - start
    
    data = read_file(get_name_from_datetime(start))
    sampwidth = data.getsampwidth()
    fmts = (None, "=B", "=h", None, "=l")
    fmt = fmts[sampwidth]
    dcs  = (None, 128, 0, None, 0)
    dc = dcs[sampwidth]

    array = np.zeros(data.getnframes() * length // duration.seconds)
    print(data.getnframes() * length // duration.seconds)

    iframe = data.readframes(data.getnframes() * (dt - start).seconds // duration.seconds)
    for i in range(data.getnframes() * length // duration.seconds):
        iframe = data.readframes(1)
        iframe = struct.unpack(fmt, iframe)[0]
        iframe -= dc
        array[i] = iframe
    d = {'max': max(array), 'min': min(array),
                              'mean': np.mean(array), 'median': np.median(array), '25': np.percentile(array, 25),
                             '75': np.percentile(array, 75), 'power': np.sqrt(np.mean(array**2))}
    fft = np.abs(np.fft.fft(array, n=10))
    for j in range(10):
        d[str(j)] = [fft[j]]
    return pd.DataFrame(data=d)

    
all_files = os.listdir('Audio')
get_audio_features_by_datetime(datetime.strptime('20181014 14:10:10', '%Y%m%d %H:%M:%S'), 1, all_files)
```
